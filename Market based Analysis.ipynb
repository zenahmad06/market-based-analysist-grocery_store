```python
import pandas as pd
data=pd.read_csv("C:/Users/User/Documents/Groceries data.csv")
data.head()
```

**Prepare dataset** : apriori requires dataset is one-hot encoded format

```python
data_prep=data.copy()
#tanda
data_prep['status']=1
#jumlah transaksai barang pengelompokan
#unstack hierarki index ke column
status_barang=data_prep.groupby(['Member_number','itemDescription'])['status'].sum().unstack().reset_index().fillna(0).set_index("Member_number")
status_barang
```

convert ke bilangan biner (lihat kolom bags)

```python
def encode (x):
    if x<=0:
        return 0
    else:
        return 1
status_enc=status_barang.applymap(encode)
status_enc
```

**find frequent items**: apply apriori algorytm in dataset

```python
from mlxtend.frequent_patterns import apriori
#min_support : threshold yang digunakan jika kurang dari threshold frequent tidak dihitung
frequent_item=apriori(status_enc,min_support=0.1,use_colnames=True)
frequent_item
```

 so based on support, there's no common item in grocery store, if there's no item common we can use both confidence or lift, meanwhile if there's common item, we can just use lift. definition support is propbability transcation item / all transactions
